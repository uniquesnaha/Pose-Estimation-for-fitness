```python
import gradio as gr
from utils.exercise_utils import Exercise
def bmicalc(name,height,weight):
    bmiv=round(float(weight)/(float(height)**2),2)
    stri=result(bmiv)
    return {bmi:gr.update(value="Hi "+name+"! Your BMI is "+str(bmiv),visible=True),
             index:gr.update(value=stri,visible=True)}

def result(bmi):
    if(bmi<=18.5):
        return "Uh-Oh! You are Underweight"
    elif(bmi<=24.9):
        return "Congratulations! You are Healthy"
    else:
        return "Oops! You are Overweight"
def exer():
    return{c1:gr.update(visible=True),c2:gr.update(visible=True),c3:gr.update(visible=True),c4:gr.update(visible=True),c5:gr.update(visible=True),
           c6:gr.update(visible=True),c7:gr.update(visible=True),c8:gr.update(visible=True),c9:gr.update(visible=False),c10:gr.update(visible=False),c11:gr.update(visible=False)}
    
def yog():
    return{c1:gr.update(visible=False),c2:gr.update(visible=False),c3:gr.update(visible=False),c4:gr.update(visible=False),c5:gr.update(visible=False),
           c6:gr.update(visible=False),c7:gr.update(visible=False),c8:gr.update(visible=False),c9:gr.update(visible=True),c10:gr.update(visible=True),c11:gr.update(visible=True)}
 
def select(option):
    pose=Exercise(1,option)
    pose.estimate_exercise()

def display_diet(recommendation,choice,vn):
    if(recommendation=="Yes"):
        if(choice=="Weight Loss"):
            if(vn=="Veg"):
                return "veg-weightloss.jpeg"
            return "nonveg-weightloss.jpeg"
        if(vn=="Veg"):
            return "veg-weight-gain.jpeg"
        return "nonveg-weight-gain.jpeg"
    return "SampleDiet.jpg"
    
    
    
    
####################################################################################################################    
with gr.Blocks(title="Pose.AI") as demo:
    with gr.Tab("Home"):
        with gr.Row():
            with gr.Row():
                name=gr.Textbox(placeholder="Enter you name",label="Name")
                with gr.Column():
                    age=gr.Textbox(label="Age")
                    gender=gr.Dropdown(["MALE","FEMALE","OTHER"],label="Gender")
                
                with gr.Row():
                    height=gr.Textbox(label="Height")
                    weight=gr.Textbox(label="Weight")
        bmi=gr.Label(color="#984063",visible=False)
        
        with gr.Row():
            calculate=gr.Button("CALCULATE BMI")
            index=gr.Label(color="#984063",visible=False)
            gr.Button.click(fn=bmicalc,inputs=[name,height,weight],outputs=[bmi,index],self=calculate)
        
        with gr.Row():
            l1=gr.Label("Pick your choice")
            exercise=gr.Button("EXERCISE")
            yoga=gr.Button("YOGA")
           
 #############################################################################################################3       

    with gr.Tab("Select an Exercise/Yoga Pose from below:"):
        with gr.Row():
            with gr.Column(visible=False) as c1:
                gr.Label("Pushup")
                e1=gr.Textbox("pushup",visible=False)
                gr.Image(value="Pushups.jpg",shape=(None,None))
                pushup=gr.Button(value="Click here to Start!")
            with gr.Column(visible=False) as c2:
                gr.Label("Plank")
                gr.Image(value="Plank.webp",shape=(None,None))
                e2=gr.Textbox("plank",visible=False)
                plank=gr.Button(value="Click here to Start!")
            with gr.Column(visible=False) as c3:
                gr.Label("Squat")
                e3=gr.Textbox("squat",visible=False)
                gr.Image(value="Squat.jpg",shape=(None,None))
                squat=gr.Button(value="Click here to Start!")
            with gr.Column(visible=False) as c4:
                gr.Label("Jumping Jack")
                e4=gr.Textbox("jumpingjack",visible=False)
                gr.Image(value="JumpingJack.webp",shape=(None,None))
                jj=gr.Button(value="Click here to Start!")
       
            with gr.Column(visible=False) as c5:
                gr.Label("Shoulder Press")
                gr.Image(value="Shoulder.webp",shape=(None,None))
                e5=gr.Textbox("shoulderpress",visible=False)
                sp=gr.Button(value="Click here to Start!")
            with gr.Column(visible=False) as c6:
                gr.Label("Single Arm Curl")
                e6=gr.Textbox("singlearmcurl",visible=False)
                gr.Image(value="1arm.jpg",shape=(None,None))
                single=gr.Button(value="Click here to Start!")
            with gr.Column(visible=False) as c7:
                gr.Label("Double Arm Curl")
                e7=gr.Textbox("doublearmcurl",visible=False)
                gr.Image(value="doublearm.png",shape=(None,None))
                doublep=gr.Button(value="Click here to Start!")
            with gr.Column(visible=False) as c8:
                gr.Label("High Knees")
                e8=gr.Textbox("highknees",visible=False)
                gr.Image(value="highknees.jpg",shape=(None,None))
                high=gr.Button(value="Click here to Start!")
            with gr.Column(visible=False) as c9:
                gr.Label("T-Pose")
                e9=gr.Textbox("tpose",visible=False)
                gr.Image(value="tpose.jpg",shape=(None,None))
                tpose=gr.Button(value="Click here to Start!")
            with gr.Column(visible=False) as c10:
                gr.Label("Tree Pose")
                e10=gr.Textbox("treepose",visible=False)
                gr.Image(value="treepose.jpg",shape=(None,None))
                tree=gr.Button(value="Click here to Start!")
            with gr.Column(visible=False) as c11:
                gr.Label("Warrior Pose")
                e11=gr.Textbox("warriorpose",visible=False)
                gr.Image(value="warrior2.jpg",shape=(None,None))
                warrior=gr.Button(value="Click here to Start!")
                gr.Button.click(fn=exer,outputs=[c1,c2,c3,c4,c5,c6,c7,c8,c9,c10,c11],self=exercise)
                gr.Button.click(fn=yog,outputs=[c1,c2,c3,c4,c5,c6,c7,c8,c9,c10,c11],self=yoga)
                gr.Button.click(fn=select,inputs=e1,self=pushup)
                gr.Button.click(fn=select,inputs=e2,self=plank)
                gr.Button.click(fn=select,inputs=e3,self=squat)
                gr.Button.click(fn=select,inputs=e4,self=jj)
                gr.Button.click(fn=select,inputs=e7,self=doublep)
                gr.Button.click(fn=select,inputs=e8,self=high)
                gr.Button.click(fn=select,inputs=e5,self=sp)
                gr.Button.click(fn=select,inputs=e6,self=single)
                gr.Button.click(fn=select,inputs=e11,self=warrior)
                gr.Button.click(fn=select,inputs=e9,self=tpose)
                gr.Button.click(fn=select,inputs=e10,self=tree)
        

   ##########################################################################################################         
    with gr.Tab("Diet"):
        with gr.Row() as r1:
            recommendation = gr.Dropdown(["Yes","No"],label="Do you need diet recommendation?")
            choice = gr.Dropdown(["Weight Loss","Weight Gain"],label="Choice")
            vn = gr.Dropdown(["Veg","Non Veg"],label="Veg/Non-Veg")
        with gr.Row():
            submit = gr.Button("SUBMIT")
        with gr.Row():
            diet_recommendation = gr.Image(value="Loading.jpg")
            gr.Button.click(fn=display_diet,inputs=[recommendation,choice,vn],outputs=diet_recommendation,self=submit)
            #gr.Image(value=diet_recommendation)#shape=(None,None))

####################################################################################################################
    with gr.Tab("Report"):
        
        with gr.Row():
            x=gr.Video(value="c:/Users/ashwi/MrPose/output.mp4",format='mp4',interactive=False)
            
            gr.Video.style(height=500,width=500,self=x)
           # gr.Video.play(fn=p_video,inputs=x,self=x)
            gr.Image(value="test.png",shape=(None,None))
            gr.Image(value="fitmama.png",shape=(None,None))
       
            
         
        
demo.launch()
```

```python
import gradio as gr

get_name = gr.Interface(lambda name: name, inputs="textbox", outputs="textbox")
prepend_hello = gr.Interface(lambda name: f"Hello {name}!", inputs="textbox", outputs="textbox")
append_nice = gr.Interface(lambda greeting: f"{greeting} Nice to meet you!",
                           inputs="textbox", outputs=gr.Textbox(label="Greeting"))
demo = gr.Series(get_name, prepend_hello, append_nice)

if __name__ == "__main__":
    demo.launch()
```
